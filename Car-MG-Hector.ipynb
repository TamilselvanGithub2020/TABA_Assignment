```python
#Importing required modules
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import os
import csv
import pandas as pd
from selenium.webdriver.common.keys import Keys
```

```python
# Create a CSV file with headers based on the attributes captured per review
csv_file = open('MGHector.csv', 'w', encoding="UTF-8", newline="")
writer = csv.writer(csv_file)
writer.writerow(['Review-Text', 'Review-Date', 'Star-Rating','Web-Site']) # define the headers
```

```python
locationOfWebdriver = "chromedriver.exe"
driver = webdriver.Chrome(locationOfWebdriver)
```

```python
# Going to cardekho.com  
# Scrape next pages one by one
for i in range(1,20,1):
    # Wait for 3 secs before going to new URL
    time.sleep(3)
    if(i==1):
        url = "https://www.cardekho.com/mg/hector/user-reviews"
    else:
        url = "https://www.cardekho.com/mg/hector/user-reviews/" + str(i)
    driver.get(url)
    try:
        
        # find all the reviews in the page
        reviewlist = driver.find_elements_by_xpath("//div[@class='readReviewBox']")
        print(len(reviewlist))
        for review in reviewlist:
            review_dict = {}  # for each job declare a dictionary

            # get the review text
            try:
                                                      #changed class name here
                revtext = review.find_element_by_xpath(".//p[@class='contentheight']")    # For title and heading
                reviewtext = revtext.text
            except:
                reviewtext = ''
            try:
                reviewertext = review.find_element_by_xpath(".//div[@class='name']")
                reviewer = reviewertext.text
            except:
                 reviewer = ''

            # get the review date
            try:
                # changed date block (split)
                reviewdate = review.find_element_by_xpath(".//div[@class='date']") 
                cusReviewdate = reviewdate.text
                review1 = cusReviewdate.split(":")
                review2 = review1[1].split("|")
                reviewdate = review2[0].strip()
            except:
                 reviewdate = ''

           # get the review date
            try:  #check for .5 rating
                rating=0
                starrating = review.find_element_by_xpath(".//div[@class='starRating']")
                r1 = starrating.find_element_by_xpath(".//span[@title='1']").get_attribute('class')
                print("Rating 1 is: ", r1)
                if("icon-star-full-fill" in r1):
                    rating = rating+1
                    r2 = starrating.find_element_by_xpath(".//span[@title='2']").get_attribute('class')
                    if("icon-star-full-fill" in r2):
                        rating = rating+1
                        r3 = starrating.find_element_by_xpath(".//span[@title='3']").get_attribute('class')
                        if("icon-star-full-fill" in r3):
                            rating = rating+1
                            r4 = starrating.find_element_by_xpath(".//span[@title='4']").get_attribute('class')
                            if("icon-star-full-fill" in r4):
                                rating = rating+1
                                r5 = starrating.find_element_by_xpath(".//span[@title='5']").get_attribute('class')
                                if("icon-star-full-fill" in r5):
                                    rating = rating +1  
                if("icon-star-full-empty" in [r1,r2,r3,r4,r5] ):
                    rating = rating + 0.5
                
            except:
                 rating = ''

            # get the review rating

            review_dict['revtext']    = reviewtext
           # review_dict['reviewer']   = reviewer
            review_dict['reviewdate'] = reviewdate
            review_dict['rating'] = rating
            review_dict['reviewsite'] = 'cardekho.com'  
            print(review_dict)

            #capture data of the job in the CSV file
            writer.writerow(review_dict.values())
    except Exception as e:
        print(e)
        csv_file.close()
       # break
```
